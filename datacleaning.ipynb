```python
import pandas as pd
```

```python
df=pd.read_csv("quarterly_results.csv")
```

```python
df_2=df.rename(columns={'Unnamed: 1': 'Metrics'})
```

```python
df_2["Metrics"].unique()
```

```python
df_2["Metrics"] = df_2["Metrics"].str.replace("\xa0+", "")
```

```python
df_2["Metrics"].unique()
```

```python
df_2=df_2[df_2["Metrics"]!='raw pdf']
```

```python
df_2
```

```python
category=pd.read_csv("top_companies.csv")
```

```python
category=category.rename(columns={'name': 'Company'})
```

```python
category
```

```python
merged_df= pd.merge(df_2, category[['sector', 'rank', 'Company','url']], on='Company', how='left')
```

```python
merged_df.to_csv("quarterly_results_cleaned.csv", index=False)
```

```python
merged_df .columns
```

```python
import pandas as pd

# Assuming df is your original DataFrame
# Step 1: Pivot all monthly columns into a single 'period' column
month_cols = [col for col in merged_df.columns if col not in ['Company', 'Metrics', 'sector', 'rank', 'url']]

df_melted = merged_df.melt(
    id_vars=['Company', 'Metrics', 'sector', 'rank', 'url'],
    value_vars=month_cols,
    var_name='period',
    value_name='value'
)

# Step 2: Convert 'period' to datetime (1st of each month)
df_melted['date'] = pd.to_datetime(df_melted['period'], format='%b %Y')

# Optional: set it to 1st of month explicitly
df_melted['date'] = df_melted['date'].dt.to_period('M').dt.to_timestamp()

# Final dataframe will have:
# ['Company', 'Metrics', 'sector', 'rank', 'url', 'period', 'value', 'date']
```

```python
df_melted=df_melted[df_melted["date"] >= "2022-01-01"]
```

```python
df_melted=df_melted[df_melted["Metrics"] != "Raw PDF"]
```

```python
df_melted.to_csv("quarterly_results_cleaned.csv", index=False)
```
